```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the Excel file path
excel_file = 'cleaned_pathways.xlsx'

# Define the list of DataFrame names
df_names = ['df_mu', 'df_md', 'df_fu', 'df_fd']

# Load each sheet into separate DataFrames
dfs = pd.read_excel(excel_file, sheet_name=None)

df_mu = list(dfs.values())[3]
df_md = list(dfs.values())[2]
df_fu = list(dfs.values())[1]
df_fd = list(dfs.values())[0]
```

```python
df_fu.shape
```

```python
df_fu['unique_fu'] = ~df_fu["Name"].isin(df_mu["Name"])
# df_fu
unique_fu = df_fu[df_fu['unique_fu']]
df_mu['unique_mu'] = ~df_mu["Name"].isin(df_fu["Name"])
# df_fu
unique_mu = df_mu[df_mu['unique_mu']]
```

```python
unique_fu.head(20)
```

```python
writer = pd.ExcelWriter('final_tables_ALL_Reordered_FINAL.xlsx')

dfs = [df_mu, df_md, df_fu, df_fd]

sheet_names = ['df_mu', 'df_md', 'df_fu', 'df_fd']

for i, df in enumerate(dfs):
    # Apply transformations to the DataFrame
    df = df.iloc[:, 1:5]
    df['Name'] = df['Name'].str.replace('KEGG_', '')
    df['Name'] = df['Name'].str.replace('_', ' ')
    df['Name'] = df['Name'].str.title()
    df["negative_log10_pval"] = np.log10(df["p-value"]) * (-1)
#     df = df.head(20)
    # Save the DataFrame to a sheet in the Excel file
    df.to_excel(writer, sheet_name=sheet_names[i], index=False)

# Save the Excel file
writer.save()
   
```

# Pathway Analysis

### Rho/Rock Pathway

```python
# Assuming df is your pandas DataFrame containing the pathway data
# and "Name" is the column you want to search

# Convert the "Name" column to lowercase for case-insensitive matching
df = df_mu

df_copy = df.copy()

# Convert the "Name" column to lowercase for case-insensitive matching
df_copy['Name'] = df_copy['Name'].str.lower()

# Define the keywords to search for
keywords = ['rho', 'rhoa', 'rhob', 'roc','rock']

# Create a boolean indexing column for matching pathway names
df_copy['KeywordMatch'] = df_copy['Name'].str.contains('|'.join(keywords))

# Filter the dataframe to retrieve the rows with matching pathway names
matching_df = df_copy[df_copy['KeywordMatch']]

print(f"there are this many rho/rock related genes: {matching_df.shape[0]}")

# Display the resulting dataframe
matching_df[matching_df["unique_mu"]]
```
